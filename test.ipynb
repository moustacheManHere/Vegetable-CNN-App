```python
s3 = boto3.resource("s3")
imgBucket = s3.Bucket("devops-ca2")
for i in vegeDB:
    path = os.path.join("temp_imgs",i)
    imgBucket.upload_file(path,i)
    print(f"Uploaded {i}")
```
